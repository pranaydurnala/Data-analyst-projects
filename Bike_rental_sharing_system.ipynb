```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

day_df =pd.read_csv(r"/content/day.csv")
print(day_df)

hour_df=pd.read_csv(r"/content/hour.csv")
print(hour_df)
```

```python
# Visualize the distribution of bike rentals across different seasons
plt.figure(figsize=(8, 5))
sns.countplot(x='season', hue = 'season',legend = False, data=day_df, palette='Set3')
plt.title('Distribution of Bike Rentals by Season')
plt.xlabel('Season')
plt.ylabel('Count')
plt.xticks(ticks=[0, 1, 2, 3], labels=['Winter', 'Spring', 'Summer', 'Fall'])
plt.show()
```

```python
# Visualize the bike rental count distribution across different hours of the day
plt.figure(figsize=(12, 6))
sns.countplot(x='hr', hue = 'hr',legend = False , data=hour_df, palette='Set2')
plt.title('Bike Rental Count Distribution by Hour of the Day')
plt.xlabel('Hour')
plt.ylabel('Count')
plt.show()
```

```python
# Distribution of bike rentals by year
plt.figure(figsize=(8, 5))
sns.countplot(x='yr', hue = 'yr',legend = False,data=day_df, palette='coolwarm')
plt.title('Bike Rentals by Year')
plt.xlabel('Year (0 = 2011, 1 = 2012)')
plt.ylabel('Count of Rentals')
plt.show()
```

```python
# Distribution of bike rentals by weekday
plt.figure(figsize=(8, 5))
sns.countplot(x='weekday', hue='weekday', data=day_df, palette='Set3', legend=False)
plt.title('Bike Rentals by Weekday')
plt.xlabel('Day of the Week (0 = Sunday, 6 = Saturday)')
plt.ylabel('Count of Rentals')
plt.show()
```

```python
# Most common weather conditions during bike rentals
plt.figure(figsize=(8, 5))
sns.countplot(x='weathersit', hue = 'weathersit',legend = False,data=day_df, palette='Set1')
plt.title('Weather Conditions During Bike Rentals')
plt.xlabel('Weather Situation')
plt.ylabel('Count of Rentals')
plt.xticks(ticks=[0, 1, 2, 3], labels=[
    'Clear/Partly Cloudy',
    'Mist/Cloudy',
    'Light Snow/Rain',
    'Heavy Rain/Snow'
])
plt.show()
```

```python
# Average temperature and feels-like temperature during bike rentals
plt.figure(figsize=(10, 6))
sns.histplot(day_df['temp'], kde=True, color='blue', label='Temperature')
sns.histplot(day_df['atemp'], kde=True, color='orange', label='Feels-like Temperature', alpha=0.7)
plt.title('Temperature vs Feels-like Temperature During Bike Rentals')
plt.xlabel('Temperature (Celsius)')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
# Humidity variation in the dataset
plt.figure(figsize=(8, 5))
sns.histplot(day_df['hum'], kde=True, color='green')
plt.title('Humidity Distribution During Bike Rentals')
plt.xlabel('Humidity (Normalized)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Wind speed variation in the dataset
plt.figure(figsize=(8, 5))
sns.histplot(day_df['windspeed'], kde=True, color='purple')
plt.title('Windspeed Distribution During Bike Rentals')
plt.xlabel('Windspeed (Normalized)')
plt.ylabel('Frequency')
plt.show()
```

```python
# This scatter plot illustrates the relationship between temperature and total bike rentals, with seasons distinguished by color.
plt.figure(figsize=(10, 6))
sns.scatterplot(x='temp', y='cnt', data=day_df, hue='season', palette='coolwarm')
plt.title('Temperature vs. Bike Rentals')
plt.xlabel('Temperature (Celsius)')
plt.ylabel('Total Bike Rentals')
plt.legend(title='Season')
plt.show()
```

```python
# This pair plot visualizes the pairwise relationships between temperature, humidity, windspeed, and total bike rentals.
sns.pairplot(day_df[['temp', 'atemp', 'hum', 'windspeed', 'cnt']])
plt.suptitle('Pairwise Relationships Between Features', y=1.02)
plt.show()
```

```python
# Bike rentals comparison between holidays and non-holidays using a box plot
plt.figure(figsize=(8, 5))
sns.boxplot(x='holiday', y='cnt', data=day_df, palette='Set2')
plt.title('Distribution of Bike Rentals on Holidays vs Non-Holidays')
plt.xlabel('Holiday (0 = No, 1 = Yes)')
plt.ylabel('Total Rentals')
plt.show()
```

```python
# Creating a violin plot to show the distribution of casual and registered users
plt.figure(figsize=(8, 5))
sns.violinplot(data=hour_df[['casual', 'registered']], palette=['blue', 'orange'])
plt.title('Distribution of Casual and Registered Users')
plt.xlabel('User Type')
plt.ylabel('Number of Users')
plt.xticks(ticks=[0, 1], labels=['Casual', 'Registered'])
plt.show()
```

```python
# Calculating the proportion of bike rentals by weather situation
# Creating a pie chart to show the proportion of bike rentals by weather situation
weather_counts = day_df['weathersit'].value_counts()
weather_labels = weather_counts.index.map({
    1: 'Clear/Partly Cloudy',
    2: 'Mist/Cloudy',
    3: 'Light Snow/Rain',
    4: 'Heavy Rain/Snow'
})
plt.figure(figsize=(8, 8))
plt.pie(weather_counts, labels=weather_labels, autopct='%1.1f%%', colors=['lightblue', 'lightgreen', 'lightcoral', 'lightgrey'], startangle=140)
plt.title('Proportion of Bike Rentals by Weather Condition')
plt.show()
```
