```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(r"/content/IMDb_Dataset.csv")
df
```

```python
df.isna().sum()
```

```python
df.info()
```

```python
df.describe()
```

```python
plt.figure(figsize=(14, 7))
sns.lineplot(data=df, x='Year', y='IMDb Rating', marker='o', color='b')
plt.title("IMDb Rating Trend Over Time")
plt.xlabel("Year")
plt.ylabel("IMDb Rating")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=df['Genre'].value_counts().values, y=df['Genre'].value_counts().index, palette='viridis')
plt.title("Genre Popularity Distribution")
plt.xlabel("Number of Movies")
plt.ylabel("Genre")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['IMDb Rating'], bins=20, kde=True, color='skyblue')
plt.title("Distribution of IMDb Ratings")
plt.xlabel("IMDb Rating")
plt.ylabel("Frequency")
plt.show()
```

```python
top_directors = df['Director'].value_counts().head(10)
plt.figure(figsize=(8, 8))
plt.pie(top_directors.values, labels=top_directors.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("Set3"))
plt.title("Top 10 Directors by Movie Count")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.regplot(data=df, x='Duration (minutes)', y='IMDb Rating', scatter_kws={'alpha':0.5}, color='purple')
plt.title("Duration vs. IMDb Rating with Regression Line")
plt.xlabel("Duration (minutes)")
plt.ylabel("IMDb Rating")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.violinplot(data=df, x='Certificates', y='IMDb Rating', palette='Set2', inner="quartile")
plt.title("Distribution of IMDb Ratings by Certification")
plt.xlabel("Certification")
plt.ylabel("IMDb Rating")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='Genre', y='Duration (minutes)', palette="Spectral")
plt.title("Distribution of Movie Duration by Genre")
plt.xlabel("Genre")
plt.ylabel("Duration (minutes)")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 8))
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
sns.heatmap(df[numeric_cols].corr(), annot=True, cmap='coolwarm', square=True)
plt.title("Correlation Heatmap of Numeric Features")
plt.show()
```

```python
movies_per_year = df['Year'].value_counts().sort_index()
plt.figure(figsize=(14, 7))
sns.barplot(x=movies_per_year.index, y=movies_per_year.values, palette="Blues")
plt.title("Number of Movies Released Each Year")
plt.xlabel("Year")
plt.ylabel("Number of Movies")
plt.xticks(rotation=45)
plt.show()
```
