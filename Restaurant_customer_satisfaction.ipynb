```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(r"/content/restaurant_customer_satisfaction.csv")
df
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
# 1. Correlation Heatmap: What factors influence customer satisfaction the most?
plt.figure(figsize=(12, 8))
corr_matrix = df.corr(numeric_only=True)  # Ensure only numeric columns are used
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap: Factors Influencing Customer Satisfaction')
plt.show()
```

```python
# 2. Bar Plot: Distribution of customer satisfaction across different demographics (Age)
plt.figure(figsize=(10, 6))
sns.barplot(x='Age', y='HighSatisfaction', hue = 'Age',legend = False,data=df, palette='viridis')
plt.title('Average Satisfaction by Age Group')
plt.ylabel('Average Satisfaction')
plt.xlabel('Age Group')
plt.show()
```

```python

# 3. Box Plot: How does visit frequency relate to customer spending and satisfaction?
plt.figure(figsize=(10, 6))
sns.boxplot(x='VisitFrequency', y='AverageSpend',hue = 'VisitFrequency',legend = False, data=df, palette='Set2')
plt.title('Distribution of Spending by Visit Frequency')
plt.ylabel('Average Spend')
plt.xlabel('Visit Frequency')
plt.show()
```

```python
# 5. Scatter Plot: Does group size impact satisfaction?
plt.figure(figsize=(10, 6))
sns.scatterplot(x='GroupSize', y='HighSatisfaction', hue='ServiceRating', data=df, palette='coolwarm', size='ServiceRating', sizes=(20, 200))
plt.title('Group Size vs Satisfaction')
plt.ylabel('Satisfaction (0=No, 1=Yes)')
plt.xlabel('Group Size')
plt.legend(title='Service Rating', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python

# 6. Line Plot: How does the time of visit affect spending?
plt.figure(figsize=(10, 6))
sns.lineplot(x='TimeOfVisit', y='AverageSpend',hue = 'TimeOfVisit',legend = False, data=df, marker='o', palette='tab10')
plt.title('Average Spending by Time of Visit')
plt.ylabel('Average Spend')
plt.xlabel('Time of Visit')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x='OnlineReservation', y='HighSatisfaction',hue = 'OnlineReservation',legend = False, data=df, palette='coolwarm')
plt.title('Satisfaction by Online Reservation')
plt.ylabel('Average Satisfaction')
plt.xlabel('Online Reservation (0=No, 1=Yes)')
plt.show()
```

```python

# 5. Violin Plot: To compare the distribution of food ratings by different preferred cuisines
plt.figure(figsize=(12, 8))
sns.violinplot(x='PreferredCuisine', y='FoodRating', hue = 'PreferredCuisine',legend = 'auto',data=df, palette='muted', split=True)
plt.title('Distribution of Food Ratings by Preferred Cuisine')
plt.ylabel('Food Rating')
plt.xlabel('Preferred Cuisine')
plt.show()
```
