```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(r"/content/drive/MyDrive/data sets/customer_purchase_data.csv")
df
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.groupby(by='Gender').count()
```

```python
min_age = df.Age.min()
print(min_age)

max_age = df.Age.max()
print(max_age)
```

```python
df.drop_duplicates()
df
```

```python
Sorting = df[df['NumberOfPurchases']>15].sort_values(by='TimeSpentOnWebsite',ascending=False).head()
Sorting
```

```python
# Get the top 5 customers who availed the most discounts
top_discount_customers = df.nlargest(5, 'DiscountsAvailed')
print(top_discount_customers)
```

```python
# Group by 'Product Category' and sum the 'PurchaseStatus' to count purchases for each category
product_purchases = df.groupby('ProductCategory')['PurchaseStatus'].sum().sort_values(ascending=False)
print(product_purchases)
```

```python
# Group by 'Product Category' and sum the 'PurchaseStatus' to count purchases for each category
product_purchases = df.groupby('ProductCategory')['PurchaseStatus'].sum().sort_values(ascending=False)
plt.figure(figsize=(6,4))
plt.pie(product_purchases, labels=product_purchases.index, autopct='%1.1f%%')
plt.title('Number of Purchases by Product Category')
plt.show()
```

```python
avg_age = df.groupby('LoyaltyProgram')['Age']
gender_distribution = df.groupby('LoyaltyProgram')['Gender'].value_counts(normalize=True)

print("Average Age by Loyalty Program Membership:")
print(avg_age)

print("\nGender Distribution by Loyalty Program Membership:")
print(gender_distribution)
```

```python
age_bins = [0, 30, 60, 100]
age_labels = ['Young (0-30)', 'Middle-aged (31-60)', 'Senior (61+)']

# Create a new column for categorized age
df['AgeCategory'] = pd.cut(df['Age'], bins=age_bins, labels=age_labels)

age_discount_distribution = df.groupby('AgeCategory')['DiscountsAvailed'].sum()

print("Total Discounts Availed by Age Category:")
print(age_discount_distribution)
```

```python
sns.countplot(data=df, x='AgeCategory', hue='Gender')
plt.title('Gender Distribution by Age Categories')
plt.xlabel('Age Category')
plt.ylabel('Count')
plt.legend(title='Gender', labels=['Male', 'Female'])
plt.show()
```

```python
# Filter the DataFrame for customers aged more than 50
age_over_50 = df[df['Age'] > 50]
# Create a histogram for the 'Number of Purchases'
plt.figure(figsize=(6, 3))
plt.hist(age_over_50['NumberOfPurchases'], bins=10, color='skyblue', edgecolor='black')
plt.title('Histogram of Number of Purchases for Customers Aged Over 50')
plt.xlabel('NumberofPurchases')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
sns.boxplot(data=df, x='AgeCategory', y='TimeSpentOnWebsite')
plt.title('Time Spent on Website by Age Category')
plt.xlabel('Age Category')
plt.ylabel('Time Spent on Website (minutes)')
plt.show()
```

```python
# Calculate the mean time spent on the website for each combination of AgeCategory and Gender
time_spent_by_gender = df.groupby(['AgeCategory', 'Gender'])['TimeSpentOnWebsite'].mean().unstack()
# Plot the stacked bar chart
time_spent_by_gender.plot(kind='bar', stacked=True, color=['orange', 'skyblue'])
plt.title('Average Time Spent on Website by Age Category and Gender')
plt.xlabel('Age Category')
plt.ylabel('Average Time Spent (minutes)')
plt.legend(title='Gender', labels=['Male', 'Female'])
plt.show()
```

```python
age_categories = df['AgeCategory'].unique()

for category in age_categories:
    plt.figure()
    gender_counts = df[df['AgeCategory'] == category]['Gender'].value_counts()
    plt.pie(gender_counts, labels=['Male', 'Female'], autopct='%1.1f%%', colors=['pink', 'skyblue'])
    plt.title(f'Gender Proportion in {category}')
    plt.show()
```

```python
sns.boxplot(data=df, x='AgeCategory', y='NumberOfPurchases', hue='Gender')
plt.title('Number of Purchases by Age Category and Gender')
plt.xlabel('Age Category')
plt.ylabel('Number of Purchases')
plt.show()
```

```python
sns.barplot(data=df, x='DiscountsAvailed', y='PurchaseStatus', errorbar=None)
plt.title('Impact of Discounts on Purchase Likelihood')
plt.xlabel('Number of Discounts Availed')
plt.ylabel('Purchase Likelihood (0 = No, 1 = Yes)')
plt.show()
```

```python
plt.figure(figsize=(7, 5))
sns.scatterplot(x='AnnualIncome', y='NumberOfPurchases', hue='PurchaseStatus', data=df, palette='coolwarm')
plt.title('AnnualIncome vs Number of Purchases')
plt.xlabel('AnnualIncome')
plt.ylabel('NumberOfPurchases')
plt.legend(title='Purchase Status', labels=['No', 'Yes'])
plt.show()
```

```python
# Pair plot for numerical features
sns.pairplot(df, hue='PurchaseStatus', vars=['Age', 'AnnualIncome', 'NumberOfPurchases', 'TimeSpentOnWebsite'])
plt.suptitle('Pair Plot of Customer Behavior Dataset', y=1.02)
plt.show()
```
